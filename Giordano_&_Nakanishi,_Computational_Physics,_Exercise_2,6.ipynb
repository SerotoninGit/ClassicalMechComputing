Problem Statement: Use the Euler method to calculate cannon shel trajectories ignoring both air
drag and the effect of air density (actually, ignoring the former automatically
rules out the latter). Compare your results with those in Figure 2.4, and with
the exact solution.

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt

# Initial Conditions
x0 = 0.0
y0 = 0.0
v0 = 1000.0
angle = 45.0 * np.pi/180.0
vx = v0 * np.cos(angle)
vy = v0 * np.sin(angle)

g = 9.81
dt = 0.01
time_max = 200.0

# Current positions start at initial
x = x0
y = y0

x_list = [x]
y_list = [y]

t = 0.0
while t < time_max and y >= 0:
    # Update velocity
    vy = vy - g * dt

    # Update positions
    x = x + vx * dt
    y = y + vy * dt

    # Record positions
    x_list.append(x)
    y_list.append(y)

    t += dt

plt.plot(x_list, y_list)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Cannon Shell Trajectory')
plt.grid(True)
plt.show()

#print max range
print("Max range: ", x_list[-1], "m")
```

Compared to figure 2.4 and the real solutions, these trajectories are longer range than the real solutions with air resistance since ignoring drag makes the projectile travel further and stay in the air longer.
