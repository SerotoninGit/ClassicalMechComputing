Problem Statement: In our model of the cannon shel trajectory we have assumed that the acceleration
due to gravity, g, is a constant. It wil, of course, depend on altitude. Add this
to the model and calculate how much it affects the range.

Thinking: Lets assume the inverse square law for gravity \g(y) = g_0(\frac{R + y}{R})


```python
#include Librarys
import numpy as np
import matplotlib.pyplot as plt

# Constants and initial conditions
R = 6.371e6     # Earth's radius (m)
g0 = 9.81        # gravitational acceleration at Earth's surface (m/s^2)
x0 = 0.0
y0 = 0.0
v0 = 1000.0
angle = 45.0 * np.pi/180.0
vx = v0 * np.cos(angle)
vy = v0 * np.sin(angle)

dt = 0.01
time_max = 200.0

x = x0
y = y0
x_list = [x]
y_list = [y]

t = 0.0
while t < time_max and y >= 0:
    # Altitude-dependent gravity, inverse square law
    g = g0 * (R/(R+y))**2

    # Update velocity
    vy = vy - g * dt

    # Update positions
    x = x + vx * dt
    y = y + vy * dt

    x_list.append(x)
    y_list.append(y)
    t += dt

# Compute range as the maximum x reached at ground level
range_with_variable_g = x_list[-1]

plt.plot(x_list, y_list, label='Trajectory with Variable g')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Cannon Shell Trajectory with Altitude-Dependent Gravity')
plt.grid(True)
plt.legend()
plt.show()

print("Range with altitude-dependent gravity:", range_with_variable_g, "m")
```

Solution: From 2.6 with the same initial conditions, the Range does not change much at all, however if we increase the initial velocity to 1000 m/s (about modern naval artilerly values, from google) we do begin to see that in the model with altitude-dependent gravity, the range begins to increase, since gravity weakens slightly as you go higher.
